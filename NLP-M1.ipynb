```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

mask_image_path = 'palestine_map.png'  
mask_image = np.array(Image.open(mask_image_path).convert('L'))

mask_image = np.where(mask_image > 128, 255, 0).astype(np.uint8)


stopwords = set(STOPWORDS)

wordcloud = WordCloud(width=800, height=800, background_color='white', mask=mask_image,
                      contour_width=1, contour_color='black', colormap='viridis',
                      max_font_size=100, random_state=42, max_words=500, font_step=1,
                      stopwords=stopwords).generate(paragraph)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.show()
```

```python


def generate_wordcloud(mask_image_path, text):
    mask_image = np.array(Image.open(mask_image_path).convert('L'))

    mask_image = np.where(mask_image > 128, 255, 0).astype(np.uint8)

    words = word_tokenize(text)

    words = [word.lower() for word in words if word.isalpha()]

    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    cleaned_text = ' '.join(words)

    wordcloud = WordCloud(width=800, height=800, background_color='white', mask=mask_image,
                        contour_width=1, contour_color='black', colormap='viridis',
                        max_font_size=100, random_state=42, max_words=500, font_step=1).generate(cleaned_text)



    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  
    plt.show()
```
