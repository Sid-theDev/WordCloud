```python
import nltk
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt 
import numpy as np
from wordcloud import WordCloud, STOPWORDS
from PIL import Image

nltk.download('punkt')
nltk.download('stopwords')

stopwords = set(STOPWORDS)

def generate_wordcloud(mask_image_path, text):
    mask_image = np.array(Image.open(mask_image_path).convert('L'))

    mask_image = np.where(mask_image > 128, 255, 0).astype(np.uint8)

    words = word_tokenize(text)

    words = [word.lower() for word in words if word.isalpha()]

    nltk_stop_words = set(nltk.corpus.stopwords.words('english'))
    words = [word for word in words if word not in nltk_stop_words]

    cleaned_text = ' '.join(words)

    wordcloud = WordCloud(width=800, height=800, background_color='white', mask=mask_image,
                          contour_width=1, contour_color='black', colormap='viridis',
                          max_font_size=100, random_state=42, max_words=500, font_step=1, stopwords=stopwords).generate(cleaned_text)

    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  
    plt.show()

text = """In the heart of Spain, where legends tread,
Real Madrid, in white and gold, is led.
Champions of Europe, kings of the field,
A legacy of glory, forever sealed.

With Santiago's roar, the Bernabu stands,
Witness to triumphs, crafted by skilled hands.
From Di Stfano's grace to Ronaldo's might,
They shine through history, stars burning bright.

Hala Madrid! The anthem we sing,
For the greatest club, where champions cling.
Through victory and honor, their banner unfurled,
Real Madrid, the greatest club in the world."""

masked_image_path = "real_madrid.jpg"

generate_wordcloud(text=text, mask_image_path=masked_image_path)
```
